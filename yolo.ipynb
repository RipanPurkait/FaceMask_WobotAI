```python
import os
import random
import shutil

def move_random_files(src_dir, dest_dir, percentage):
    # Ensure the destination directory exists
    os.makedirs(dest_dir, exist_ok=True)

    # List all files in the source directory
    files = os.listdir(src_dir)

    # Calculate the number of files to move
    num_files_to_move = int(len(files) * (percentage / 100))

    # Randomly select files to move
    files_to_move = random.sample(files, num_files_to_move)

    # Move the files
    for file in files_to_move:
        src_path = os.path.join(src_dir, file)
        dest_path = os.path.join(dest_dir, file)
        shutil.move(src_path, dest_path)

    print(f"Moved {num_files_to_move} files from {src_dir} to {dest_dir}.")

# Paths
source_directory = '/content/drive/MyDrive/FaceMask/data/train/face_with_mask'  # Replace with your source directory path
destination_directory = '/content/drive/MyDrive/FaceMask/data/val/face_with_mask'  # Replace with your destination directory path
percentage_to_move = 30

# Move 20% of the files randomly
move_random_files(source_directory, destination_directory, percentage_to_move)
```
